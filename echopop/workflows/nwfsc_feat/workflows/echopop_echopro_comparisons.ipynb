# Compare EchoPro and Echopop report outputs

This notebook ingests the reports generated by Echopop and compares them with those produced by EchoPro. It is assumed that the EchoPro and Echopop reports are either in separate folders, or have unique names/identifiers that differentiate them. Each comparison function can optionally produce a plot that renders in the environment (or notebook if not within a script) and save to a specified directory.

## Importing required libraries and modules

```python
from pathlib import Path
from echopop.workflows.nwfsc_feat.workflows import comparisons as comp
```

## Entering parameters

This code snippet is mainly for convenience and is not necessary to generate the associated figures/outputs.

This compares the aged-length haul counts reports.

```python
# Aged length haul counts
echopro_aged_length_haul_counts = comp.read_pivot_table_report(
    ECHOPRO_ROOT / ECHOPRO_TO_ECHOPOP_FILE_MAP["aged_length_haul_counts"]["echopro"]
)
echopop_aged_length_haul_counts = comp.read_pivot_table_report(
    ECHOPOP_ROOT / ECHOPRO_TO_ECHOPOP_FILE_MAP["aged_length_haul_counts"]["echopop"]
)
comp.plot_haul_count_comparisons(
    echopro=echopro_aged_length_haul_counts, 
    echopop=echopop_aged_length_haul_counts,
    save_filepath=SAVE_DIRECTORY / "aged_length_haul_counts.png"
)
```

## Transect population estimates

This compares the transect population estimates reports. The `comparisions.plot_geodata` function allows for a dictionary of entries for `save_filepath` that can be useful when there are *many* parameters in a given file, such as the transect population estimates. In this case, we can divy up all of the estimates into five separate figures: 

1. Abundance
2. Number density
3. Biomass
4. Biomass density
5. NASC

!!! COLORS BACKWARDS FOR DIFFERENCE

```python
# Transect population estimates
echopro_transect_estimates = comp.read_geodata(
    filepath=ECHOPRO_ROOT / ECHOPRO_TO_ECHOPOP_FILE_MAP["transect_results_full"]["echopro"],
)

echopop_transect_estimates = comp.read_geodata(
    filepath=ECHOPOP_ROOT / ECHOPRO_TO_ECHOPOP_FILE_MAP["transect_results_full"]["echopop"],
)

comp.plot_geodata(
    echopro=echopro_transect_estimates,
    echopop=echopop_transect_estimates,
    save_filepath={
        ("abundance", "abundance_male", "abundance_female"): 
            SAVE_DIRECTORY / "transect_abundances.png",
        ("number_density", "number_density_male", "number_density_female"):
            SAVE_DIRECTORY / "transect_number_densities.png",
        ("biomass", "biomass_male", "biomass_female"):
            SAVE_DIRECTORY / "transect_biomasses.png",
        ("biomass_density", "biomass_density_male", "biomass_density_female"):
            SAVE_DIRECTORY / "transect_biomass_densities.png",
        "nasc": SAVE_DIRECTORY / "transect_nasc.png",
    }
)
```
